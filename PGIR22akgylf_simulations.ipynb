# PGIR22akgylf Nova Lightcurve Simulation and Period Recovery Test

## Overview

This notebook performs a comprehensive validation of our period detection methodology using a combination of real TESS observations and controlled simulations. The goal is to demonstrate that our detrending pipeline can reliably recover known periodic signals even when they are embedded in complex astrophysical trends from real nova eruption data. The key point to illustrate is that the original signal is recovered after detredning. The presence of the trend does not significantly alter the frequenc/period of the injected signal.

## Methodology

### Step-by-Step Validation Process

**Step 1: Extract Real Nova Lightcurve**
- Downloads TESS Sector 55 full-frame images covering nova PGIR22akgylf
- Applies 3×3 pixel square aperture photometry with background subtraction
- Uses the same extraction methodology as our main analysis

**Step 2: Analyze Original Data and Save Trend Model**
- Runs complete period search on the real nova lightcurve
- Applies Savitzky-Golay detrending (window=101, polyorder=5) 
- **Critically**: Saves the extracted trend model for later reuse
- Estimates the amplitude of real periodic variability for simulation scaling

**Step 3: Create Controlled Simulation**
- Generates a synthetic lightcurve with a known 0.200-day periodic signal
- Uses the same time sampling as the real TESS observations
- Sets periodic amplitude to match the real variability amplitude
- Adds signal to a constant baseline flux level

**Step 4: Validate Pure Signal Recovery**
- Tests period detection on the clean simulated lightcurve (no trend)
- Confirms our pipeline can detect the 0.200-day period accurately
- Establishes baseline performance without astrophysical complications

**Step 5: Add Real Complexity**
- Combines the 0.200-day simulation with the **real trend** saved from Step 2
- Creates a hybrid lightcurve: known periodic signal + authentic nova evolution
- Preserves all the complex time-variable behavior of the actual eruption

**Step 6: Ultimate Validation Test**
- Runs period search on the trend+simulation combination
- Tests whether detrending can recover the known 0.200-day period
- Measures detection accuracy and validates methodology robustness

## Scientific Significance

### Why This Test Matters

1. **Real-World Validation**: Uses authentic astrophysical trends rather than idealized mathematical functions
2. **Controlled Experiment**: We know the true period (0.200 days), enabling precise accuracy measurement  
3. **Worst-Case Scenario**: Nova eruptions produce some of the most complex lightcurve evolution patterns
4. **Parameter Matching**: Simulation amplitude matches real variability, ensuring realistic signal-to-noise conditions

### Expected Outcomes

- **Success Criterion**: Recovery of 0.200-day period with <0.1% error
- **Validation**: Confirms our detrending parameters are appropriate for nova-like variability
- **Confidence**: Demonstrates reliability for detecting similar periods in other nova eruption data

## Technical Details

- **Detrending Method**: Savitzky-Golay filter (subtraction-based)
- **Period Search**: Lomb-Scargle periodogram (0.05-3.0 day range)
- **Outlier Removal**: 3-σ clipping before and after detrending
- **Phase Analysis**: Sine wave fitting with epoch determination
- **Amplitude Matching**: Simulated signal scaled to real variability level

## Output Products

The notebook generates:
- Individual lightcurve plots for each processing step
- Periodograms showing detected periods
- Phased lightcurves with fitted sine waves  
- Comparison plot showing original, simulated, and combined data
- Quantitative accuracy metrics and validation results

This comprehensive test provides strong evidence for the reliability of our period detection methodology when applied to nova eruption lightcurves with complex temporal evolution.

```python
# PGIR22 Nova Lightcurve Simulation and Period Recovery Test
# 
# This notebook performs a comprehensive test of period detection capabilities using:
# 1. Real TESS Sector 55 lightcurve of nova PGIR22akgylf
# 2. Simulated 0.200-day periodic signal
# 3. Combination of simulated signal with real trend
#
# The goal is to validate that our detrending methodology can recover
# known periodic signals even in the presence of complex trends from real data.

import os
import lightkurve as lk
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter
from scipy.optimize import curve_fit
import astropy.units as u
%matplotlib inline

# Reference time for nova eruption
t_0 = 2459807.6900  # Reference JD for nova eruption

# Monkey patch the LightCurve.plot method to use reference time
import lightkurve as lk
import matplotlib.pyplot as plt

# Store the original plot method
_original_plot = lk.LightCurve.plot

def plot_with_reference_time(self, *args, **kwargs):
    """Modified plot method that uses reference time t_0"""
    # Create a copy to avoid modifying the original
    lc_copy = self.copy()
    
    # Convert time to days relative to t_0 (only if time is in JD format)
    if hasattr(lc_copy.time, 'jd'):
        time_rel = lc_copy.time.jd - t_0
        lc_copy.time = time_rel
        
        # Call the original plot method
        result = _original_plot(lc_copy, *args, **kwargs)
        
        # Set the x-axis label to show reference time
        #plt.xlabel(f'JD - {t_0} (days)')
        plt.xlabel(f'Days since $t_0$ (days)')
        
        
        return result
    else:
        # If not JD format, use original method
        return _original_plot(self, *args, **kwargs)

# Replace the plot method
lk.LightCurve.plot = plot_with_reference_time

def simple_detrend_and_period_search_with_model_save(lc, period_range=(0.05, 3.0), titleprefixstring="", detrend_method="subtraction"):
    """
    Modified version that saves the detrending model for later reuse.
    
    Returns:
    tuple: (best_period, jd0, detrend_model)
        - best_period: The detected period in days
        - jd0: The epoch of minimum in JD
        - detrend_model: The trend that was removed (for reuse in simulations)
    """
    import numpy as np
    import matplotlib.pyplot as plt
    
    # Step 1: Clip the lightcurve at 3-sigma
    print(f"About to plot sigma-clipped LC with {len(lc)} -> {len(lc.remove_outliers(sigma=3))} points")
    clipped_lc = lc.remove_outliers(sigma=3)
    clipped_lc.plot(lw=2, marker='.', linestyle='None')
    plt.title(f"{titleprefixstring} sigma-clipped (N={len(clipped_lc)})")
    plt.show()
    
    # Step 2: Detrend the lightcurve and SAVE the model
    if detrend_method == "division":
        # Original method: using scipy's Savitzky-Golay filter wrapped in lc.flatten() (division-based)
        detrended_lc = clipped_lc.flatten(window_length=101, polyorder=5)
        # For division method, the trend is flux/detrended_flux
        detrend_model = clipped_lc.flux.value / detrended_lc.flux.value
        detrend_type = "detrended (division)"
    elif detrend_method == "subtraction":
        # New method: manual Savitzky-Golay filter and subtraction
        from scipy.signal import savgol_filter
        from lightkurve import LightCurve
        
        # Get the time and flux arrays
        time = clipped_lc.time.value
        flux = clipped_lc.flux.value
        
        # Compute the trend using the same parameters as in flatten()
        trend = savgol_filter(flux, window_length=101, polyorder=5)
        
        # Save the trend model
        detrend_model = trend
        
        # Create a new lightcurve with the detrended flux (subtraction)
        detrended_lc = LightCurve(time=clipped_lc.time, flux=flux - trend)
        detrend_type = "detrended (subtraction)"
    else:
        raise ValueError("detrend_method must be either 'division' or 'subtraction'")
    
    # Plot the detrended lightcurve
    print(f"About to plot detrended LC with {len(detrended_lc)} points")
    detrended_lc.plot(lw=2, marker='.', linestyle='None')
    plt.title(f"{titleprefixstring} sigma-clipped, {detrend_type} (N={len(detrended_lc)})")
    plt.show()
    
    # Step 2.5: Clip the lightcurve again
    print(f"About to plot second sigma-clipped LC with {len(detrended_lc)} -> {len(detrended_lc.remove_outliers(sigma=3))} points")
    detrended_lc = detrended_lc.remove_outliers(sigma=3)
    detrended_lc.plot(lw=2, marker='.', linestyle='None')
    plt.title(f"{titleprefixstring} sigma-clipped, {detrend_type}, sigma-clipped (N={len(detrended_lc)})")
    plt.show()    
    
    # Step 3: Perform Lomb-Scargle periodogram analysis
    min_period, max_period = period_range
    min_frequency = 1 / max_period
    max_frequency = 1 / min_period
    
    # Create the periodogram
    periodogram = detrended_lc.to_periodogram(method="lombscargle", 
                                             minimum_frequency=min_frequency, 
                                             maximum_frequency=max_frequency)
    
    # Find and print the best period
    best_period = periodogram.period_at_max_power
    print(f"The best period is: {best_period:.4f} days")
    
    # Step 4: Plot the periodogram as a function of frequency
    plt.figure(figsize=(10, 5))
    plt.plot(periodogram.frequency.value, periodogram.power, label="Lomb-Scargle Power")
    plt.axvline(x=1/best_period.value, color="red", linestyle="--", label=f"Best Period = {best_period:.4f} days")
    plt.xlabel("Frequency (1/day)")
    plt.ylabel("Power")
    plt.title(f"{titleprefixstring} Lomb-Scargle Periodogram")
    plt.legend()
    plt.grid()
    plt.show()
    
    # Step 5: Create phased lightcurve plot with sine wave fit to determine phase 0
    from scipy.optimize import curve_fit
    
    time_values = detrended_lc.time.value
    flux_values = detrended_lc.flux.value
    
    # Calculate initial phases
    phases = (time_values - time_values[0]) / best_period.value % 1
    
    # Define sine wave function for fitting
    def sine_wave(phase, amplitude, phase_offset, vertical_offset):
        return amplitude * np.sin(2 * np.pi * (phase - phase_offset)) + vertical_offset
    
    # Initial guess for parameters
    amplitude_guess = (np.max(flux_values) - np.min(flux_values)) / 2
    vertical_offset_guess = np.mean(flux_values)
    phase_offset_guess = 0
    
    # Fit sine wave to the data
    try:
        popt, pcov = curve_fit(sine_wave, phases, flux_values, 
                              p0=[amplitude_guess, phase_offset_guess, vertical_offset_guess],
                              maxfev=5000)
        amplitude_fit, phase_offset_fit, vertical_offset_fit = popt
        
        # For sin(2π(φ - φ₀)), minimum occurs when 2π(φ - φ₀) = -π/2 + 2πn
        # So φ_min = φ₀ - 1/4 (taking n=0 and adjusting to [0,1))
        phase_of_minimum = (phase_offset_fit - 0.25) % 1
        
        # If amplitude is negative, the function is inverted (max becomes min)
        if amplitude_fit < 0:
            amplitude_fit = -amplitude_fit
            phase_of_minimum = (phase_of_minimum + 0.5) % 1  # Shift by half period
        
        print(f"Fitted sine wave parameters:")
        print(f"  Amplitude: {amplitude_fit:.6f}")
        print(f"  Phase offset: {phase_offset_fit:.6f}")
        print(f"  Vertical offset: {vertical_offset_fit:.6f}")
        print(f"  Phase of minimum: {phase_of_minimum:.6f}")
        
        # Adjust phases so fitted minimum occurs at phase 0
        adjusted_phases = (phases - phase_of_minimum) % 1
        
        # Calculate JD0 (epoch of minimum)
        jd0 = 2457000 + time_values[0] + phase_of_minimum * best_period.value
        
        # Update the phase offset for plotting the fitted curve
        adjusted_phase_offset = (phase_offset_fit - phase_of_minimum) % 1
        
    except Exception as e:
        print(f"Sine wave fitting failed: {e}")
        print("Using simple minimum finding instead...")
        # Fallback to simple minimum finding
        min_flux_idx = np.argmin(flux_values)
        phase_of_minimum = phases[min_flux_idx]
        adjusted_phases = (phases - phase_of_minimum) % 1
        
        # Calculate JD0 using simple minimum
        jd0 = 2457000 + time_values[min_flux_idx]
        
        # Set dummy values for plotting
        amplitude_fit, adjusted_phase_offset, vertical_offset_fit = amplitude_guess, 0, vertical_offset_guess
    
    # Print light elements
    print(f"Light Elements:")
    print(f"  Period: {best_period:.6f} days")
    print(f"  JD0 (epoch of minimum): {jd0:.6f}")
    
    # Create the phased lightcurve plot from -0.5 to 1.5
    plt.figure(figsize=(12, 7))
    
    # Create extended phase arrays for plotting from -0.5 to 1.5
    extended_phases = []
    extended_flux = []
    
    for phase_shift in [-1, 0, 1]:
        extended_phases.extend(adjusted_phases + phase_shift)
        extended_flux.extend(flux_values)
    
    extended_phases = np.array(extended_phases)
    extended_flux = np.array(extended_flux)
    
    # Filter to plot range -0.5 to 1.5
    plot_mask = (extended_phases >= -0.5) & (extended_phases <= 1.5)
    plot_phases = extended_phases[plot_mask]
    plot_flux = extended_flux[plot_mask]
    
    # Plot the data points
    plt.scatter(plot_phases, plot_flux, alpha=0.6, s=20, color='tab:blue', label='')
    
    # Create binned version for cleaner visualization
    phase_bins = np.linspace(-0.5, 1.5, 20)
    binned_flux = []
    binned_flux_err = []
    bin_centers = []
    
    for i in range(len(phase_bins)-1):
        mask = (plot_phases >= phase_bins[i]) & (plot_phases < phase_bins[i+1])
        if np.sum(mask) > 0:
            binned_flux.append(np.mean(plot_flux[mask]))
            binned_flux_err.append(np.std(plot_flux[mask]) / np.sqrt(np.sum(mask)))
            bin_centers.append((phase_bins[i] + phase_bins[i+1]) / 2)
    
    bin_centers = np.array(bin_centers)
    binned_flux = np.array(binned_flux)
    binned_flux_err = np.array(binned_flux_err)
    
    # Plot binned data
    if len(bin_centers) > 0:
        plt.errorbar(bin_centers, binned_flux, yerr=binned_flux_err, 
                    color='tab:red', marker='o', markersize=4, linewidth=2, 
                    label='Binned', capsize=3, linestyle='None')
    
    # Plot the fitted sine wave
    phase_fine = np.linspace(-0.5, 1.5, 500)
    try:
        # Plot the fitted curve in the adjusted phase coordinate system
        # The sine wave should have its minimum at phase 0 in the adjusted coordinates
        fitted_curve = amplitude_fit * np.sin(2 * np.pi * (phase_fine - 0.25)) + vertical_offset_fit
        plt.plot(phase_fine, fitted_curve, 'g-', linewidth=2, label='Fitted sine wave')
    except:
        pass
    
    # Add vertical lines at key phases
    plt.axvline(x=0, color='green', linestyle='--', alpha=0.7, label='Phase 0')
    plt.axvline(x=-1, color='green', linestyle='--', alpha=0.5)
    plt.axvline(x=1, color='green', linestyle='--', alpha=0.5)
    
    plt.xlabel('Phase')
    plt.ylabel('Flux')
    plt.title(f'{titleprefixstring} Phased Lightcurve (Period = {best_period:.4f} days, JD0 = {jd0:.6f})')
    plt.xlim(-0.5, 1.5)
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.show()
    
    return best_period.value, jd0, detrend_model

def simple_detrend_and_period_search(lc, period_range=(0.05, 3.0), titleprefixstring="", detrend_method="subtraction"):
    """
    Regular version without model saving for consistency with original code.
    """
    best_period, jd0, _ = simple_detrend_and_period_search_with_model_save(
        lc, period_range, titleprefixstring, detrend_method)
    return best_period, jd0

# Create output directory
output_dir = "pgir22_simulation_test"
os.makedirs(output_dir, exist_ok=True)

print("="*80)
print("STEP 1: Extract TESS Sector 55 lightcurve of PGIR22akgylf nova")
print("="*80)

# Set target source name and position
target_source_name = "PGIR22akgylf"
search_lc = lk.search_tesscut("20:00:29.254 +34:53:09.17")
print(f"Available TESS sectors for {target_source_name}:")
print(search_lc)

# Download Sector 55 (index 3 based on the original notebook)
print(f"\nDownloading TESS Sector 55 cutout...")
tpf_cutout = search_lc[3].download(cutout_size=12, quality_bitmask='hard')

# Define target aperture mask (3x3 centered at 5:8, 5:8 as in original)
target_source_aperture_mask = np.zeros((tpf_cutout.shape[1:]), dtype='bool')
target_source_aperture_mask[5:8, 5:8] = True

# Extract the background-subtracted lightcurve
custom_background_exclusion_mask = np.zeros((tpf_cutout.shape[1:]), dtype='bool')
custom_background_exclusion_mask[4:9, 4:9] = True  # extended target region
custom_background_exclusion_mask[2:5, 6:9] = True  # ATO J300.1356+34.8776 region
custom_background_exclusion_mask = custom_background_exclusion_mask | target_source_aperture_mask

background_mask = ~tpf_cutout.create_threshold_mask(threshold=0.001, reference_pixel=None) * ~custom_background_exclusion_mask

# Extract lightcurves
uncorrected_lc = tpf_cutout.to_lightcurve(aperture_mask=target_source_aperture_mask)
n_target_pixels = target_source_aperture_mask.sum()
n_background_pixels = background_mask.sum()
background_lc_per_pixel = tpf_cutout.to_lightcurve(aperture_mask=background_mask) / n_background_pixels
background_estimate_lc = background_lc_per_pixel * n_target_pixels
bkg_subtracted_lc = uncorrected_lc - background_estimate_lc.flux

# Plot the background-subtracted lightcurve
plt.figure(figsize=(12, 6))
bkg_subtracted_lc.plot(lw=2, marker='.', markersize=3, linestyle='None')
plt.title(f"{target_source_name}, TESS Sector {tpf_cutout.sector}, background-subtracted lightcurve")
plt.savefig(os.path.join(output_dir, f"step1_original_lightcurve.png"), format="png", dpi=300)
plt.show()

print(f"Extracted lightcurve with {len(bkg_subtracted_lc)} data points")
print(f"Time range: {bkg_subtracted_lc.time.value.min():.3f} to {bkg_subtracted_lc.time.value.max():.3f} TJD")
print(f"Flux range: {bkg_subtracted_lc.flux.value.min():.1f} to {bkg_subtracted_lc.flux.value.max():.1f}")

print("\n" + "="*80)
print("STEP 2: Run period search and save the detrending model")
print("="*80)

print("Running period search on original lightcurve...")
original_period, original_jd0, detrend_model = simple_detrend_and_period_search_with_model_save(
    bkg_subtracted_lc, period_range=(0.05, 3.0), 
    titleprefixstring=f"{target_source_name} original:", 
    detrend_method="subtraction"
)

print(f"\nOriginal lightcurve analysis results:")
print(f"  Detected period: {original_period:.6f} days")
print(f"  Epoch (JD0): {original_jd0:.6f}")
print(f"  Detrend model saved with {len(detrend_model)} points")

# Estimate the amplitude of the real periodic variability for comparison
# We'll use the detrended lightcurve to estimate this
clipped_lc = bkg_subtracted_lc.remove_outliers(sigma=3)
time_values = clipped_lc.time.value
flux_values = clipped_lc.flux.value
trend = savgol_filter(flux_values, window_length=101, polyorder=5)
detrended_flux = flux_values - trend
real_periodic_amplitude = np.std(detrended_flux)
print(f"  Estimated periodic variability amplitude: {real_periodic_amplitude:.3f} flux units")

print("\n" + "="*80)
print("STEP 3: Create simulated lightcurve with 0.200-day period")
print("="*80)

# Use the same time sampling as the original lightcurve
time_sim = bkg_subtracted_lc.time.value
median_flux = np.median(bkg_subtracted_lc.flux.value)

# Create a 0.200-day periodic signal with comparable amplitude to real variability
sim_period = 0.200  # days
sim_amplitude = real_periodic_amplitude  # Use same amplitude as real periodic signal
sim_frequency = 1.0 / sim_period
sim_phase = 0.0

print(f"Creating simulated lightcurve:")
print(f"  Period: {sim_period} days")
print(f"  Amplitude: {sim_amplitude:.3f} flux units (matching real variability)")
print(f"  Baseline flux: {median_flux:.1f}")

# Calculate sine wave values
sine_wave = sim_amplitude * np.sin(2.0 * np.pi * sim_frequency * time_sim + sim_phase)

# Create constant baseline + sine wave
simulated_flux = np.ones_like(time_sim) * median_flux + sine_wave

# Create LightCurve object
simulated_lc = lk.LightCurve(
    time=bkg_subtracted_lc.time,
    flux=simulated_flux * bkg_subtracted_lc.flux.unit
)

# Plot the simulated lightcurve
plt.figure(figsize=(12, 6))
simulated_lc.plot(lw=2, marker='.', markersize=3, linestyle='None')
plt.title(f"Simulated Lightcurve: Constant + 0.200-day sine wave")
plt.savefig(os.path.join(output_dir, f"step3_simulated_lightcurve.png"), format="png", dpi=300)
plt.show()

print("\n" + "="*80)
print("STEP 4: Test period recovery on pure simulated lightcurve")
print("="*80)

print("Running period search on simulated lightcurve (no trend)...")
sim_period_detected, sim_jd0, _ = simple_detrend_and_period_search_with_model_save(
    simulated_lc, period_range=(0.05, 3.0), 
    titleprefixstring=f"Simulated (0.200d):", 
    detrend_method="subtraction"
)

print(f"\nSimulated lightcurve analysis results:")
print(f"  True period: {sim_period} days")
print(f"  Detected period: {sim_period_detected:.6f} days")
print(f"  Period error: {abs(sim_period_detected - sim_period):.6f} days")
print(f"  Relative error: {abs(sim_period_detected - sim_period)/sim_period*100:.3f}%")

if abs(sim_period_detected - sim_period) < 0.001:
    print("  ✓ Period recovery successful!")
else:
    print("  ✗ Period recovery failed!")

print("\n" + "="*80)
print("STEP 5: Add original trend to simulated lightcurve")
print("="*80)

print("Adding the saved detrending model to the simulated lightcurve...")

# Add the trend model to the simulated flux
# The detrend_model contains the trend that was subtracted from the original data
trend_plus_sim_flux = simulated_flux + detrend_model

# Create new LightCurve object
trend_plus_sim_lc = lk.LightCurve(
    time=bkg_subtracted_lc.time,
    flux=trend_plus_sim_flux * bkg_subtracted_lc.flux.unit
)

# Plot the trend + simulated lightcurve
plt.figure(figsize=(12, 6))
trend_plus_sim_lc.plot(lw=2, marker='.', markersize=3, linestyle='None')
plt.title(f"Simulated Lightcurve + Original Trend: 0.200-day signal with real nova trend")
plt.savefig(os.path.join(output_dir, f"step5_trend_plus_simulated.png"), format="png", dpi=300)
plt.show()

print(f"Created trend+simulation lightcurve:")
print(f"  Number of points: {len(trend_plus_sim_lc)}")
print(f"  Flux range: {trend_plus_sim_lc.flux.value.min():.1f} to {trend_plus_sim_lc.flux.value.max():.1f}")
print(f"  Original trend amplitude: {(detrend_model.max() - detrend_model.min()):.1f}")
print(f"  Simulated signal amplitude: {sim_amplitude:.3f}")

print("\n" + "="*80)
print("STEP 6: Test period recovery with trend + simulation")
print("="*80)

print("Running period search on trend + simulated lightcurve...")
print(f"About to run period search on trend+sim LC with {len(trend_plus_sim_lc)} points")
print(f"Flux range: {trend_plus_sim_lc.flux.value.min():.1f} to {trend_plus_sim_lc.flux.value.max():.1f}")

final_period_detected, final_jd0, _ = simple_detrend_and_period_search_with_model_save(
    trend_plus_sim_lc, period_range=(0.05, 3.0), 
    titleprefixstring=f"Trend + Sim (0.200d):", 
    detrend_method="subtraction"
)

print(f"\nFinal analysis results (trend + simulation):")
print(f"  True period: {sim_period} days")
print(f"  Detected period: {final_period_detected:.6f} days")
print(f"  Period error: {abs(final_period_detected - sim_period):.6f} days")
print(f"  Relative error: {abs(final_period_detected - sim_period)/sim_period*100:.3f}%")

if abs(final_period_detected - sim_period) < 0.001:
    print("  ✓ Period recovery with trend successful!")
else:
    print("  ✗ Period recovery with trend failed!")

print("Completed Step 6 period search.")

print("\n" + "="*80)
print("SUMMARY OF RESULTS")
print("="*80)

print(f"Original PGIR22akgylf lightcurve:")
print(f"  Detected period: {original_period:.6f} days")
print(f"  Periodic amplitude: {real_periodic_amplitude:.3f} flux units")

print(f"\nSimulated 0.200-day signal (no trend):")
print(f"  True period: {sim_period} days")
print(f"  Detected period: {sim_period_detected:.6f} days")
print(f"  Recovery accuracy: {abs(sim_period_detected - sim_period)/sim_period*100:.3f}% error")

print(f"\nSimulated 0.200-day signal + original trend:")
print(f"  True period: {sim_period} days")
print(f"  Detected period: {final_period_detected:.6f} days")
print(f"  Recovery accuracy: {abs(final_period_detected - sim_period)/sim_period*100:.3f}% error")

# Create comparison plot
print("\nCreating final comparison plot...")
fig, axes = plt.subplots(3, 1, figsize=(15, 10))

# Plot 1: Original lightcurve
print("Plotting original lightcurve on subplot 1...")
axes[0].plot(bkg_subtracted_lc.time.value, bkg_subtracted_lc.flux.value, '.', markersize=2)
axes[0].set_title(f"Original PGIR22akgylf Lightcurve (Period: {original_period:.4f} days)")
axes[0].set_ylabel('Flux')
axes[0].grid(True, alpha=0.3)

# Plot 2: Simulated lightcurve (no trend)  
print("Plotting simulated lightcurve on subplot 2...")
axes[1].plot(simulated_lc.time.value, simulated_lc.flux.value, '.', markersize=2)
axes[1].set_title(f"Simulated 0.200-day Signal (Detected: {sim_period_detected:.4f} days)")
axes[1].set_ylabel('Flux')
axes[1].grid(True, alpha=0.3)

# Plot 3: Simulated + trend
print("Plotting trend+simulated lightcurve on subplot 3...")
axes[2].plot(trend_plus_sim_lc.time.value, trend_plus_sim_lc.flux.value, '.', markersize=2)
axes[2].set_title(f"Simulated 0.200-day + Original Trend (Detected: {final_period_detected:.4f} days)")
axes[2].set_ylabel('Flux')
axes[2].set_xlabel('Time (TJD)')
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(os.path.join(output_dir, f"summary_comparison.png"), format="png", dpi=300)
plt.show()
print("Final comparison plot completed.")

print(f"\nAll plots saved to: {output_dir}/")
print("\n" + "="*80)
print("ANALYSIS COMPLETE")
print("="*80)

# Final validation
if abs(final_period_detected - sim_period) < 0.001:
    print("\n✓ SUCCESS: The detrending methodology successfully recovered the")
    print(f"  0.200-day period even in the presence of the complex trend from")
    print(f"  the real nova lightcurve!")
else:
    print("\n✗ WARNING: Period recovery was not successful. This may indicate")
    print(f"  that the detrending parameters need adjustment or that the")
    print(f"  simulated signal amplitude is too small relative to the trend.")

print(f"\nThis validates that our period detection pipeline can reliably")
print(f"identify periodic signals with amplitudes of ~{sim_amplitude:.3f} flux units")
print(f"in the presence of complex astrophysical trends.")
```
